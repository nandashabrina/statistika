```python
#soal poin c : menghitung nilai mean
import numpy as np

dummyBiggest5 = [159264, 43944, 34656, 32700, 29400]
avg_dummyBiggest5 = np.mean(dummyBiggest5)
print(avg_dummyBiggest5)
```

```python
#soal poin c : menghitung nilai median
import numpy as np

dummyBiggest5 = [159264, 43944, 34656, 32700, 29400]
med_dummyBiggest5 = np.median(dummyBiggest5)
dummyBiggest5.sort()
dummyBiggest5_sort = ' '.join(map(str, dummyBiggest5))
print('Urutan produk dummy dengan total order 5 terbesar : {:s}'.format(dummyBiggest5_sort))
print(med_dummyBiggest5)
```

```python
#soal poin c : menghitung nilai kuartile dan IQR
import numpy as np

dummy_data = np.array([29400, 32700, 34656, 43944, 159264])
Q3 = np.quantile(dummy_data, 0.75)
Q1 = np.quantile(dummy_data, 0.25)
IQR = Q3 - Q1

print("Q1 of data is : %s" %(Q1))
print("Q3 of data is : %s" %(Q3))
print("IQR value is  : %s" %(IQR))
```

```python
#soal poin d: menghitung variasi dan standar deviasi
import statistics
import pandas as pd
import numpy as np

dataset = [29400, 32700, 34656, 43944, 159264]
variance = np.var(dataset)
sdev = statistics.stdev(dataset)
print("Value of variance is : %s" %(variance))
print("Value of standard deviation is : %s" %(sdev))
```
